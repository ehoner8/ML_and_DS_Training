```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
axes[0][1].plot(x, y)
fig.subplots_adjust(wspace=0.9, hspace=0.5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x, label="X vs X")
ax.plot(x, x**2, label="X vs X^2", marker="D", markersize=5, ls="--", markerfacecolor="blue")
ax.legend()
```
