```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df
```

```python
plt.figure(figsize=(12,4), dpi=125)
sns.scatterplot(x="salary", y="sales", data=df, size="work experience", hue="level of education")
plt.savefig("myfirstseaplot.png")
```

```python
sns.rugplot(x="salary", data=df, height = 0.5)
```

```python
sns.set(style="whitegrid")
sns.displot(data=df, x="salary", bins=15, edgecolor="purple", lw=2)
```

```python
sns.histplot(data=df, x="salary", kde=True)
```

```python
sns.kdeplot(data=df, x="salary")
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=["Age"])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages, x="Age")
```

```python
sns.kdeplot(data=sample_ages, x="Age", clip=[0, 100], bw_adjust=0.1, fill=True)
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x="level of education", hue="division", palette="Set1")
```

```python
plt.figure(figsize=(10, 4), dpi=130)
sns.barplot(data=df, x="level of education", y="salary", estimator=np.mean, ci="sd", palette="Set1", hue="division")
plt.legend(loc=(1.1, 0.5))
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.jointplot(data=df, x="math score", y="reading score", hue="gender")
```

```python
sns.pairplot(data=df, hue="gender", corner=True)
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="box", row="lunch", col="test preparation course", palette="Set1")
```

```python
g = sns.PairGrid(data=df, hue="gender")
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.histplot)
g = g.add_legend()
```
